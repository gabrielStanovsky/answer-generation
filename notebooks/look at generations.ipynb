```python
from csv import DictReader
import random
```

```python
random.seed(0)
lines = [line for line in DictReader(open('../merge_predictions/sampled_predictions/mcscript.csv'))]
random.shuffle(lines)
```

```python
for line in lines[:100]:
    print(line['context1'] + '\n')
    for i in range(1, 11):
        print('\t'+line['question' + str(i)])
        print('\t'+line['reference' + str(i)])
        print('\t'+line['candidate' + str(i)])
        print('\t'+line['source' + str(i)])
        print()
    print('='*100)
```
